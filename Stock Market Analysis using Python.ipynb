```python
# Lets plot an OHLC chart (Open, High, Low, Close prices). We are going to use the Plotly library for the OHLC chart.

pip install plotly
```

```python
# OHLC chart for Tesla stock.


import plotly.graph_objects as go

fig = go.Figure(
    data=go.Ohlc(
        x=data.index,
        open=data["Open"],
        high=data["High"],
        low=data["Low"],
        close=data["Close"],
     )
)

fig.update_layout(
     xaxis_title="OHLC chart with Tesla Stock prices"

)  


fig.show()
```

```python
# Import matplotlib for plotting

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Plot adjusted close price data

data['Adj Close'].plot(figsize=(10, 7))

plt.title("Adjusted Close Price of Tesla", fontsize=16)

plt.ylabel('Price', fontsize=14)

plt.xlabel('Year', fontsize=14)

plt.grid(which="major", color='k', linestyle='-.', linewidth=0.5)

plt.show()


```

```python
# Historical view of the total volume of Tesla stock traded each day

data['Volume'].plot(legend=True,figsize=(10,4))

plt.title("Volume of Tesla Stock Traded Each Day", fontsize=16)

plt.grid(which="major", color='k', linestyle='-.', linewidth=0.5)

plt.show()
```

```python
#Calculation to find he percent change for each day of Tesla stock

data['Daily Return'] = data['Adj Close'].pct_change()

```
